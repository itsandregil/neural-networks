```python
import numpy as np
import nnfs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    """
    This is a linear function, this means
    that the graph is straight line.

    Linear functions are define by:
    y = mx + b, where m is the slope and b the bias
    """
    return 2*x
```

```python
def g(x):
    """
    This is a non linear function, this means
    the graph is a curve.
    """
    return 2*x**2
```

```python
# Create the data for the plots
x = np.arange(0, 5, 1e-3)

plt.plot(x, f(x)) # Plot the linear function
plt.plot(x, g(x)) # Plot the nonlinear function
plt.show()
```
