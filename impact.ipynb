### Chapter 7: Derivatives

This concept of **derivate** will let us undertands the **"impact"** of the parameters
(weights and bias) in the loss function.

```python
import numpy as np
import nnfs
import matplotlib.pyplot as plt
%matplotlib inline

# Configs
plt.rcParams["figure.figsize"] = (5, 3)
```

```python
def linear(x):
    """
    This is a linear function, this means
    that the graph is straight line.

    Linear functions are define by:
    y = mx + b, where m is the slope and b the bias
    """
    return 2*x
```

```python
def cuadratic(x):
    """
    This is a non linear function, this means
    the graph is a curve.
    """
    return 2*x**2
```

```python
# Create the data for the plots
x = np.arange(0, 5, 1e-3)

plt.plot(x, linear(x)) # Plot the linear function
plt.show()
```

### The Slope

Another way to describe the impact in a function. 
For instance, in a linear function the impact that x has on y is define by

$$slope(m) = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$$

To calculate the slope with need to points of the function.

```python
delta_y = linear(x[1]) - linear(x[0])
delta_x = x[1] - x[0]

slope = delta_y / delta_x 
print(f"Impact of x on y: {slope}")
```

For nonlinear functions (like a cuadratic function) we need to approach this in a different way.

We need to measure the slope of a **tangent line** at x, this will give us the slope,
which is the **derivative**.

The tangent line is created by drawing a line between to points that are really close in a curve, but
the curve must be differentiable at the derivation point (this means it is continuous and smooth).

For curves, there is no single slope. The slope depends on where we measure it.


```python
# Plot a curve (cuadratic function)
plt.plot(x, cuadratic(x))
plt.show()
```

```python
# The difference of each point
p2_delta = 1e-4
x1 = 2
x2 = x1 + p2_delta

# y values for x1, x2 in the cuadratic function
y1 = cuadratic(x1)
y2 = cuadratic(x2)

print((x1, y1), (x2, y2))
```

```python
# Calculate the approximate derivative and the y-intercept
# These values are important to calculate the tangent line
approximate_derivative = (y2 - y1) / (x2 - x1)
b = y2 - approximate_derivative * x2 # Intercept at (x2, y2)
```

```python
def tanget_line(x):
    """
    Calculates the value for a input in
    the tangent line.

    The Tangent Line is the line ...
    """
    return approximate_derivative * x + b
```

```python
to_plot = [x1-0.9, x1, x1+0.9] # Points for the tangent line plot

# Plot the tangent line in the cuadratic function
plt.plot(x, cuadratic(x))
plt.plot(to_plot, [tanget_line(i) for i in to_plot])

print("Approx. Derivative for f(x)", f"Where x = {x1} is {approximate_derivative}")
plt.show()
```

```python
def analytical_derivative(x):
    return 4*x

def tangent_line_analytical(x):
    return 8*x - 8

# If I know the analytical derivative of the function
# I can find the function of the tangent line, code it and
# should get the same tangent line from above.
plt.plot(x, cuadratic(x))
plt.plot(to_plot, [tangent_line_analytical(i) for i in to_plot])

print("Analytical Derivative for f(x)", f"Where x = {x1} is {analytical_derivative(x1)}")
plt.show()
```
